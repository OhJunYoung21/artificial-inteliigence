```python
import pandas as pd
import matplotlib.pyplot as plt

f = pd.read_csv('/content/train.csv')

train_1 = pd.DataFrame(f)

train_1.columns
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import  StandardScaler
import numpy as np

feature = np.array(train_1['wind_direction'])
target = train_1['high_temp']

feature = feature[:,np.newaxis]
#target = target[:,np.newaxis]

x_train,x_test,y_train,y_test = train_test_split(feature,target,test_size=0.3,random_state=0)


Lin = LinearRegression()

Lin.fit(x_train,y_train)


score = Lin.score(x_test,y_test)

print("선형회귀로 예측한 점수는 {:.2f}입니다".format(score))   #형편없는 점수가 나왔다.






```
