```python
words = ['the fool does think he is wise',
         'but the wise knows himself to be a fool']

from sklearn.feature_extraction.text import CountVectorizer

vect = CountVectorizer()  #단어를 cpu가 알아들을 수 있도록 바꿔주는 역할.
vect.fit(words)  #CountVectorizer에 words를 학습시켜준다.fit(학습)

print(len(vect.vocabulary_))
print(vect.vocabulary_)
```

```python
BOW = vect.transform(words)
print(repr(BOW))

print(BOW.toarray())
```
